Project Description:

This project aims to analyze the frequency of the word "gais" in the posts made by the X account N_CathyJKT48. The dataset consists of posts from Cathy's X account, and the goal is to count how many times the word "gais" appears in her posts. Additionally, the project calculates the total number of posts made by Cathy on X.

```python
#Import library
import pandas as pd
```

```python
#File path link

file_path1 = 'https://raw.githubusercontent.com/zackyalgiffari/CathyGaisCountDataset/main/CathyGais%208%20Maret%202024.csv'
file_path2 = 'https://raw.githubusercontent.com/zackyalgiffari/CathyGaisCountDataset/main/CathyGaiss%208%20Maret%202024.csv'
file_path3 = 'https://raw.githubusercontent.com/zackyalgiffari/CathyGaisCountDataset/main/CathyGaisss%208%20Maret%202024.csv'

# read first csv file
df1 = pd.read_csv(file_path1, delimiter=";")

# read second csv file
df2 = pd.read_csv(file_path2, delimiter=";")

# read third csv file
df3 = pd.read_csv(file_path3, delimiter=";")

# Combine dataframe
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# Save combined data to csv file
combined_df.to_csv('combined_file.csv', index=False)
```

```python
#Display head of combined data
combined_df.head()
```

```python
# Display information of combined data
combined_df.info()
```

```python
# Check for duplicated values based on 'id_str' column
duplicated_values = combined_df[combined_df.duplicated(subset=['id_str'])]
print(duplicated_values)
```

```python
# Drop duplicated values
combined_df = combined_df.drop_duplicates()
combined_df.info()
```

```python
# Perform feature engineering: convert 'full_text' to lowercase
combined_df['full_text'] = combined_df['full_text'].apply(lambda x: x.lower())

# Select specific columns for further analysis
selected_columns = ['created_at', 'id_str', 'full_text','tweet_url']
df = combined_df[selected_columns]
df.head()
```

```python
# Display information of selected data
df.info()
```

```python
# Save selected data to Excel file
df.to_excel('FinalCathyGaisCounting.xlsx', index=False)
```
